#### 4.3 Multivariate analysis using pieplot

```python
import matplotlib.pyplot as plt

# set overall figure size
plt.rcParams['figure.figsize'] = (30, 12)

# create 15 axes array
fig, axes = plt.subplots(1, 5)

# 1) Gender pie
sizes = df['gender'].value_counts()
axes[0].pie(
    sizes,
    labels=sizes.index.str.title(),    # ['Female','Male']
    autopct='%.2f%%'
)
axes[0].set_title('Gender')

# 2) Race/Ethnicity pie (sorted to AE)
order = ['group A','group B','group C','group D','group E']
sizes = df['race_ethnicity'].value_counts().reindex(order)
axes[1].pie(
    sizes,
    labels=[s.title() for s in sizes.index],
    autopct='%.2f%%'
)
axes[1].set_title('Race/Ethnicity')

# 3) Lunch pie
sizes = df['lunch'].value_counts()
axes[2].pie(
    sizes,
    labels=sizes.index.str.replace('/', ' ').str.title(),  # e.g. 'Free/Reduced'  'Free Reduced'
    autopct='%.2f%%'
)
axes[2].set_title('Lunch')

# 4) Test preparation course pie
sizes = df['test_preparation_course'].value_counts()
axes[3].pie(
    sizes,
    labels=sizes.index.str.title(),  # ['None','Completed']
    autopct='%.2f%%'
)
axes[3].set_title('Test Preparation')

# 5) Parental education pie
sizes = df['parental_level_of_education'].value_counts()
axes[4].pie(
    sizes,
    labels=[s.title() for s in sizes.index],
    autopct='%.2f%%'
)
axes[4].set_title('Parental Education')

plt.tight_layout()
plt.show()
```

#####  Insights
- Number of Male and Female students is almost equal
- Number students are greatest in Group C
- Number of students who have standard lunch are greater
- Number of students who have not enrolled in any test preparation course is greater
- Number of students whose parental education is "Some College" is greater followed closely by "Associate's Degree"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1) Load & prepare
df = pd.read_csv('data/stud.csv')  # adjust path as needed
df['average'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)

# 2) Group by gender, keeping only numeric cols
gender_group = df.groupby('gender')[['math_score', 'average']].mean()

# 3) Pull out the values for plotting
# Note: use .loc[...] rather than integer indexing
male_scores   = [gender_group.loc['male',   'average'],   gender_group.loc['male',   'math_score']]
female_scores = [gender_group.loc['female', 'average'],   gender_group.loc['female', 'math_score']]

# 4) Plot
plt.figure(figsize=(10, 8))

labels = ['Total Average', 'Math Average']
x = np.arange(len(labels))

plt.bar(x - 0.2, male_scores,   width=0.4, label='Male')
plt.bar(x + 0.2, female_scores, width=0.4, label='Female')

plt.xticks(x, labels)
plt.ylabel("Marks")
plt.title("Total Average vs Math Average by Gender", fontweight='bold')
plt.legend()
plt.tight_layout()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
